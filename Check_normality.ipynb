**Q-Q plots** - to quickly check data normality.

Normal distribution on this graphs will be a diagonal from left bottom corner to right top corner. If you want to check normlity strictly, use shapiro-wilk criterion.

*Example 1*

```python
import statsmodels.api as sm
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

data = stats.lognorm(s=1).rvs(500)
data_log = np.log(data)

ax1 = sm.qqplot(data, loc=data.mean(), scale=data.std())
ax2 = sm.qqplot(data_log, loc=data_log.mean(), scale=data_log.std())
```

*Example 2*

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
stats.probplot(data, dist="norm", plot=plt)
plt.subplot(2,2,2)
stats.probplot(data_log, dist="norm", plot=plt)
plt.show()
```

**Shapiro-Wilk normality test**

```python
print("Shapiro-Wilk normality test, W-statistic: %f, p-value: %f" % stats.shapiro(data_log))
```

**Conclusions:**
    
  * if the data has a specific distribution (Bin, Poisson) - use tests for it!
  * if not, check data normality on *Q-Q plots*. If it seems to be normal, use Student's tests, they are not very sensitive to not "verynormal" distributions.
  * If you don't want to use Student's tests and consider some sensetive for normality test, check normality with *Shapiro-Wilk normality test*.
